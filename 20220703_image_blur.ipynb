```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
plt.imshow(img)
```

```python
f = np.array([1/10]*10)
for i in range(img.shape[0]):
    for j in range(img.shape[1]-9):
        for k in range(img.shape[2]):
            img[i, j, k] = np.sum(img[i, j:j+10, k] * f)

plt.imshow(img)
```
