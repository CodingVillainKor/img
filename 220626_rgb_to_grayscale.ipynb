```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
plt.imshow(img)
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        m = (img[i, j, 0] + img[i, j, 1] + img[i, j, 2]) / 3
        img[i, j] = [m, m, m]
plt.imshow(img)
```
