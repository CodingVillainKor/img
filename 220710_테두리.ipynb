```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
```

```python
f = np.array([-1, 1])
for i in range(img.shape[0]):
    for j in range(img.shape[1]-1):
        for k in range(img.shape[2]):
            img[i, j, k] = np.abs(np.sum(img[i, j:j+2, k] * f))
        img[i, j] = img[i, j].max()

plt.imshow(img)
```

```python
img = img[:, :-1, :] / (np.abs(img[:, :-1, :]).max()+0.00001)
plt.imshow(img)
```
