```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
```

```python
f = np.ones([10, 10]) / 100
for i in range(img.shape[0]-9):
    for j in range(img.shape[1]-9):
        for k in range(img.shape[2]):
            img[i, j, k] = np.sum(img[i:i+10, j:j+10, k] * f)

plt.imshow(img)
```

```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
```

```python
f = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
for i in range(img.shape[0]-2):
    for j in range(img.shape[1]-2):
        for k in range(img.shape[2]):
            img[i, j, k] = np.abs(np.sum(img[i:i+3, j:j+3, k] * f))

img = img / img.max()
plt.imshow(img)
```
